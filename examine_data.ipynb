```python
import datetime
import parser

co2s = []
times = []
with open("co2_1551153479.68.txt", "r") as f:
    lines = f.readlines()
    line = lines[0]
    dateparts = [int(a) for a in line.split("(")[1].split(")")[0].split(", ")]
    starttime = datetime.datetime(*dateparts)
    for idx, line in enumerate(lines[1:]):
#         print(line)

        loc = line.find("'co2': ")
        
        if loc > 0:
            co2 = int(line[loc+7:loc+12].split(',')[0])
#             print(co2)
            co2s.append(co2)
            
            dateparts = [int(a) for a in line.split("(")[1].split(")")[0].split(", ")]
#             print(dateparts)
#             thistime = (datetime.datetime(*dateparts) - starttime).seconds / 3600
#             times.append(thistime)
            times.append(datetime.datetime(*dateparts))
        else:
            print("missing co2 value at position " + str(idx))
print("loaded CO2 values")
print(times)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.dates as mdates

co2s = np.array(co2s)
times = np.array(times)
fig, ax = plt.subplots()
ax.set_xlabel("Time")
ax.set_ylabel("CO2 [ppm]")
ax.plot(times, co2s)
myFmt = mdates.DateFormatter('%H:%M')
ax.xaxis.set_major_formatter(myFmt)
```
